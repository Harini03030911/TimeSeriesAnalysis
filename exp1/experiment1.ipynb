```python
import pandas as pd 
 
passenger = pd.read_csv('Electric_Production.csv') 
passenger['DATE'] = pd.to_datetime(passenger['DATE'])  
passenger.sort_values(by=['DATE'], inplace=True, ascending=True)
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
from matplotlib.pyplot import figure 
import matplotlib.pyplot as plt 
 
figure(figsize=(12, 5), dpi=80, linewidth=10) 
plt.plot(passenger['DATE'] ,passenger['IPG2211A2N']) 
plt.title('Electric Production Raw Data with Missing Values') 
plt.xlabel('Year', fontsize=14) 
plt.ylabel('Number of products', fontsize=14) 
plt.show()
```

```python
print("\nMissing Values:\n", passenger.isnull().sum())
```

```python
plt.figure(figsize=(10, 6))
plt.plot(passenger.index, passenger['IPG2211A2N'], label='Passenger Count')
plt.title('Electronic Production Data')
plt.xlabel('DATE')
plt.ylabel('Number of Products')
plt.legend()
plt.grid()
plt.show()
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler(feature_range=(0, 1))
passenger['Normalized_Passengers'] = scaler.fit_transform(passenger[['IPG2211A2N']])
```
